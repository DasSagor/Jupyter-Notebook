```python
import pandas as pd
```

```python
dataset = pd.read_csv('loan_application.csv')
dataset.head(30)
```

```python
dataset.isnull().sum()
```

```python
dataset['Gender'].fillna(dataset['Gender'].mode()[0], inplace=True)
```

```python
dataset['Married'] = dataset['Married'].fillna(dataset['Married'].mode()[0])
```

```python
en_data = dataset[['Gender','Married']]
```

```python
pd.get_dummies(en_data)
```

```python

pd.get_dummies(en_data).info()
```

```python
# One-hot encoding with Sklearn
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit_transform(en_data).toarray()
```

```python
ar = encoder.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=['Gender_Female','Gender_Male','Married_No','Married_Yes'])
```

```python
encoder = OneHotEncoder(drop='first')
ar2 = encoder.fit_transform(en_data).toarray()
ar2
```

```python
pd.DataFrame(ar2,columns=['Gender_Male','Married_Yes'])
```
