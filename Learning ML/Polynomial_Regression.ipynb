```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('polynomial.csv')
dataset.head(6)
```

```python
dataset.corr()
```

```python
plt.scatter(dataset['level'], dataset['salary'], color='red')
plt.xlabel('level')
plt.ylabel('salary')
plt.show()
```

```python
x = dataset[['level']]
y = dataset['salary']
```

```python
pf = PolynomialFeatures(degree=2)
# degree=2 is the degree of the polynomial. 2 mane x^2 porjonto calculation korbe
pf.fit(x)
x = pf.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
plt.scatter(dataset['level'], dataset['salary'], color='red')
plt.plot(dataset['level'], lr.predict(x), color='blue')
plt.xlabel('level')
plt.ylabel('salary')
plt.legend(['org', 'Predicted'])
plt.show()
```
