```python
import pandas as pd
```

```python
# Ordinal Encoding with Sklearn
df = pd.DataFrame({'size':['s', 'm', 'l', 'xl','s','xl','m','s','l','m','s','xl','m']})
df
```

```python
ord_data = [['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
df['size_en'] = oe.fit_transform(df[['size']])
df
```

```python
# Ordinal Encoding with Map function
ord_data1 = {'s': 0, 'm': 1, 'l': 2, 'xl': 3}
df['size_en1']= df['size'].map(ord_data1)
df
```

```python
ord_data1 = {'s': 3, 'm': 1, 'l': 0, 'xl': 8}
df['size_en1']= df['size'].map(ord_data1)
df
```

```python
# Ordinal Encoding with Big Dataset
dataset = pd.read_csv('loan_application.csv')
dataset.head(12)
```

```python
dataset['Property_Area'].unique()
```

```python
dataset['Property_Area'] = dataset['Property_Area'].fillna(dataset['Property_Area'].mode()[0])
dataset['Property_Area'].head(10)
```

```python
en_data_ord = [['Rural', 'Semiurban','Urban']]

from sklearn.preprocessing import OrdinalEncoder
Ordinal_Encoding = OrdinalEncoder(categories=en_data_ord)
dataset['Property_Area_En'] = Ordinal_Encoding.fit_transform(dataset[['Property_Area']])
dataset.head(50)
```
