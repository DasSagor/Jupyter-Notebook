```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=300,noise=0.05)
```

```python
df = {'data1':x[:,0], 'data2':x[:,1], 'output':y}
dataset = pd.DataFrame(df)
dataset1 = dataset.iloc[:,:-1]
```

```python
sns.scatterplot(x='data1',y='data2',data=dataset1)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3,min_samples=5)
dataset1['Predict'] = db.fit_predict(dataset1)
```

```python
sns.scatterplot(x='data1',y='data2',data=dataset1,hue='Predict')
plt.show()
```

```python
sns.scatterplot(x='data1',y='data2',data=dataset1,hue=y)
plt.show()
```
