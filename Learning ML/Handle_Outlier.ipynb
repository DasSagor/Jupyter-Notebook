```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('loan_application.csv')
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# Outlier handle with boxplot
sns.boxplot(x = 'Applicant_Income', data=dataset)
plt.show()
```

```python
sns.boxplot(x = 'Coapplicant_Income', data=dataset)
plt.show()
```

```python
# Outlier handle with distplot
sns.displot(dataset['Applicant_Income'], kde=True)
plt.show()
```
