```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('package.csv')
dataset.head(6)
```

```python
sns.scatterplot(x=dataset['CGPA'], y=dataset['Package'], data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[['CGPA']]
y = dataset['Package']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)

# random_state er upor data accuracy depend kore. Random_state er value change korle accuracy change hote pare.
```

```python
lr = LinearRegression()
# LinearRegression y = mx + c condition e kaj kore.
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
# 8.15	7.27
```

```python
lr.predict([[8.15]])
```
