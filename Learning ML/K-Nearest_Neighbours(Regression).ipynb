```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Synthetic_Employee_Data.csv')
dataset.head()
```

```python
sns.scatterplot(x='Age', y='Experience',data=dataset, hue='Salary')
plt.show()
```

```python
x = dataset.drop(columns='Salary')
y = dataset['Salary']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

```

```python
from sklearn.neighbors import KNeighborsRegressor

knc = KNeighborsRegressor(n_neighbors=7)
knc.fit(x_train,y_train)
```

```python
knc.score(x_test,y_test)*100, knc.score(x_train,y_train)*100
```
