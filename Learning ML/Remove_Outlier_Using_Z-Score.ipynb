```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('loan_application.csv')
```

```python
z_score= (dataset['Applicant_Income'] - dataset['Applicant_Income'].mean()) / (dataset['Applicant_Income'].std())
dataset['z_score'] = z_score    
new_dataset = dataset[dataset['z_score'] < 3]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x = 'Applicant_Income', data=new_dataset)
plt.show()
```
