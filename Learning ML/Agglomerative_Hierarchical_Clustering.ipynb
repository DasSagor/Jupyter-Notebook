```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('irish_flowers_raw.csv')
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(80,100))
sc.dendrogram(sc.linkage(dataset, method='single',metric='euclidean'))
plt.savefig('Agglomerative_Hierarchical_Clustering.png')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=2,linkage='single')
dataset['Predict'] = ac.fit_predict(dataset)
dataset.head()
```

```python
sns.pairplot(data=dataset,hue='Predict')
plt.show()
```
